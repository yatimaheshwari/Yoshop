```python
import pandas as pd
from bs4 import BeautifulSoup
from urllib.request import urlopen
import re
```

## Write a python programm to find product page where Products Image missing.

```python
htmdata= urlopen('https://yoshops.com')
soup = BeautifulSoup(htmdata,'html.parser')
category = soup.find_all("li", class_ = 'dropdown')
category_link = []
for link in category:
    category_link.append("https://yoshops.com" + link.a.get('href'))
all_web_pages = []
for i in category_link:
    urls = urlopen(i)
    data = BeautifulSoup(urls,'html.parser')
    for page in data.find_all("li" , class_="num"):
        all_web_pages.append(str(i)+ page.a.get('href'))
product_pages = []
for i in all_web_pages:
    urls = urlopen(i)
    sou = BeautifulSoup(urls,'html.parser')
    for link in sou.find_all("div", class_="col-sm-3 col-xs-6"):
        if link.div.a.div.div.img['src'] == '//onlinestore.wsimg.com/assets/noimage/product-5fec99477aebb10bac85d82665ec1497de4536cda3279e59089555c45cf589fa.png':
            product_pages.append("https://yoshops.com" + link.div.a.get('href'))
product_pages
```

```python
class Yo_missing_image_Product:
    def __init__(self, Web_URL ,Product_Name, Product_Details):
        self.Web_URL = Web_URL
        self.Products_Name = Product_Name
        self.Product_Details = Product_Details
object = []
for pro_page in product_pages:
    page_url = urlopen(pro_page)
    soup_page = BeautifulSoup(page_url,'html.parser')
    for i in soup_page.find_all("div" , class_= 'col-sm-5'):
        object.append(Yo_missing_image_Product(str(pro_page),i.h1.text,i.div.div.div.get('data-description')).__dict__)
dataframe = pd.DataFrame(object)
dataframe
        
    
```

```python
# Create csv file
dataframe.to_csv('List_of_Products_of_missing_Image.csv')
```
