```python
import pandas as pd
from bs4 import BeautifulSoup
from urllib.request import urlopen
import urllib.request
from PIL import Image ,ImageDraw ,ImageFont ,ImageOps
import pyqrcode
import png
from pyqrcode import QRCode
from pathlib import Path
import numpy as np
```

## Create python program for generating banner, video, blog word file,post Text file

```python
html_data = urlopen('https://yoshops.com/products/data-science-basic-training-program-for-everyone-age-10-t0-60')
soup = BeautifulSoup(html_data,'html.parser')
image_1 = soup.find_all("div",class_='col-sm-8')
image_logo = []
for i in image_1:
    image_logo.append(i.span.a.img['src'])
image_2 = soup.find_all("div" , class_= 'col-sm-4 single-product-img-col')
# imag_2_logo = []
for i in image_2:
    image_logo.append(i.div.img['src'])

# print(imag_2_logo)
title = soup.find_all("div" , class_ = 'col-sm-5')
for i in title:
    image_logo.append(i.h1.text)
discription_ = soup.find_all("p", class_ ='MsoNormal')

for i in discription_:
    if i.font != None and i.font.b != None: 
        image_logo.append(i.font.b.text)
    if i.font != None and i.font.span != None:
        image_logo.append(i.font.span.text)
print(image_logo)
```

```python
# String which represents the QR code
s = 'https://yoshops.com/products/data-science-basic-training-program-for-everyone-age-10-t0-60'
  
# Generate QR code
url = pyqrcode.create(s)
  
# Create and save the svg file naming "myqr.svg"
url.svg("myqr.svg", scale = 8)
  
# Create and save the png file naming "myqr.png"
url.png('myqr.png', scale = 6)
```

```python

assets = 'asset'
LOGO_1_IMG = Path(assets, 'banner_logo.png')
LOGO_2_IMG = Path(assets, 'banner_logo2.png')
QR_IMG = Path(assets,'myqr.png')
banner_width =750
banner_height = 1097
banner_size = (banner_width,banner_height)
WHITE = (255, 255, 255)
image = Image.new('RGB', banner_size, WHITE)
# add first image 
logo_1 = Image.open(LOGO_1_IMG)
logo_1_margin = (150,15)
image.paste(logo_1, logo_1_margin)

# add second image
logo_2 = Image.open(LOGO_2_IMG)
logo_2_margin = (150,200)
image.paste(logo_2, logo_2_margin)

# add qrcode image
qr_img = Image.open(QR_IMG)
size= (230,230)
qr_img.thumbnail(size)
qr_img_size = (500,780)
image.paste(qr_img,qr_img_size)


# add text
BLACK = (0, 0, 0)
TEXT_FONT_TYPE = Path(assets, 'Amble-Bold.ttf')
TEXT_SIZE = 24
TEXT_PADDING_HOR = 25
TEXT_PADDING_VERT = 40
IMG_TEXT = 'Data Science and Python Training Program \n for Everyone(Age=10yrs to 70yrs) \n Apply Now Offer Price Rs.899 only. \n Use Coupon Code= intern899 \n Training Program Detail: \n Course Fee : Rs.999 \n Course Duration : 1 Month= 20 classes \n Timings : Monday to Friday \n Training Modes : Online & Classroom '


draw = ImageDraw.Draw(image)
font = ImageFont.truetype(str(TEXT_FONT_TYPE), TEXT_SIZE)
offset_text = (TEXT_PADDING_HOR,700 + TEXT_PADDING_VERT)
draw.text(offset_text, IMG_TEXT, BLACK, font=font)
banner = ImageOps.expand(image,border=20,fill='teal')
banner.save('banner.png')
```
