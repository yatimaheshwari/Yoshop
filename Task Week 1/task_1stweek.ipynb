##  1.Python program to Create a excel file 

```python
import xlsxwriter
Gadget = (
    ['laptop',250000 ],
    ['mobile',15000],
    ['headphones',500],
    ['tablest',50000],
)

workbook = xlsxwriter.Workbook('Gadget.xlsx')
worksheet = workbook.add_worksheet()
row = 0
col = 0
for Gadget_list, Price in (Gadget):
    worksheet.write(row+1,col,Gadget_list)
    worksheet.write(row+1,col+1,Price)
    row+=1
# Add a bold format to use to highlight cells.
bold = workbook.add_format({'bold': 1})
# Adjust the column width.
worksheet.set_column(1, 1, 15)
worksheet.write('A1', 'Gadget_list', bold)
worksheet.write('B1', 'Price', bold)

worksheet.write(5,0,'Total Price')
worksheet.write(5,1,'=SUM(B2:B5)')

# workbook.close()
```

```python
cell_format = workbook.add_format({'font_color': 'red','bg_color': 'yellow'})
cell_frmt = workbook.add_format({'font_color': 'red','bg_color': 'green'})
row = 0
for Gadget_list, Price in (Gadget):
     worksheet.write(row+1, 0, Gadget_list,cell_format)
     worksheet.write(row+1, 1, Gadget_list,cell_frmt)
     row+=1
workbook.close()
```

## 4.Python program for Prepare Pie Chart and Bar Chart Weekly, Monthly and Yearly Reports on Yoshops Survey

```python
import pandas as pd 
yoshop_suvey_df = pd.read_excel('Yoshops Survey_1786_Updated_13_March.xlsx')
df = yoshop_suvey_df.copy()
df.head(2)
```

```python
print(df.shape)
df.info()
```

```python
type(df['Submitted Time'])
```

```python
df['Submitted Time'] = pd.to_datetime(df['Submitted Time'],errors='coerce')
df.head(2)
```

```python
df['months'] = df['Submitted Time'].dt.month
df['years'] = df['Submitted Time'].dt.year
df['weeks'] = df['Submitted Time'].dt.weekday
df.head(2)
```

```python
df['month-year']=df['months'].astype(str) + '-' + df['years'].astype(str)
df.head(2)
```
