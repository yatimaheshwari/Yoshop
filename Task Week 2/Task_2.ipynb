# 1.Analysis of Reviews given by Customers

```python
df_R.columns.values
```

```python
df_rat = df_R[df_R['stars'].notna()]
df_rat['stars'].value_counts()
```

# 7. Analysis of Number of Orders Per Month Per Year

```python
print(type(df_O['Order Date and Time Stamp']))
df_O['Order Date and Time Stamp'] = pd.to_datetime(df_O['Order Date and Time Stamp'],errors='coerce')
df_O.head(2)
```

```python
df_D_T=df_O.groupby([df_O['Order Date and Time Stamp'].dt.year.rename('Year'),df_O['Order Date and Time Stamp'].dt.month.rename('Month')])['Order #'].count().reset_index()
df_D_T.head(5)
```

```python
df_D_T['Year'].value_counts().plot(kind='bar',figsize=(14,6))
plt.title('Number of orders per year')
```

```python
year_2016 = df_D_T[(df_D_T['Year'] == 2016)]
year_2017 = df_D_T[(df_D_T['Year'] == 2017)]
year_2018 = df_D_T[(df_D_T['Year'] == 2018)]
year_2019 = df_D_T[(df_D_T['Year'] == 2019)]
year_2020 = df_D_T[(df_D_T['Year'] == 2020)]
sns.barplot(data=year_2016, x="Month", y="Order #")
plt.title('Review for year 2016')
```

```python
fig,(ax1,ax2) = plt.subplots(1,2,figsize= (18,6))
sns.barplot(data=year_2017, x="Month", y="Order #",ax=ax1)
sns.barplot(data=year_2018, x="Month", y="Order #",ax=ax2)
ax1.set_title('Review for year2017')
ax2.set_title('Review for year2018')
```

```python
fig,(ax1,ax2) = plt.subplots(1,2,figsize= (18,6))
sns.barplot(data=year_2019, x="Month", y="Order #",ax=ax1)
sns.barplot(data=year_2020, x="Month", y="Order #",ax=ax2)
ax1.set_title('Review for year2019')
ax2.set_title('Review for year2020')
```
